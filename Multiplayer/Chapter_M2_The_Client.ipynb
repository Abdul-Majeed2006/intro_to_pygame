In M1, we built a **Server** that listens.
Now we build a **Client** that speaks.

The Client's job is simple:
1.  Dial the IP (localhost).
2.  Knock on the Port (5555).
3.  Say Hello.

```python
import socket

SERVER_IP = "localhost"
PORT = 5555

client = socket.socket(socket.AF_INET, socket.SOCK_STREAM)

# 1. Connect
try:
    client.connect((SERVER_IP, PORT))
    print("Connected to Server!")
    
    # 2. Receive the Welcome Message
    # We receive 2048 bytes (more than enough for text)
    message = client.recv(2048).decode()
    print(f"Server said: {message}")
    
except socket.error as e:
    print(f"Could not connect: {e}")
```

## 2. Sending Data

Once connected, we can send strings back and forth using `send()` and `recv()`.

```python
def send_msg(text):
    client.send(str.encode(text))
    response = client.recv(2048).decode()
    print(f"Sent: {text} | Got: {response}")

# Example Usage:
# send_msg("Hello Server!")
```
