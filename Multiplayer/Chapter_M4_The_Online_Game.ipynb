We have a Server. We have a Client. We can send Objects.
Now... let's make a game.

**The Logic:**
1.  **Client**: Sends `(x, y)` to Server.
2.  **Server**: Updates `players[id] = (x, y)`.
3.  **Server**: Sends back the WHOLE `players` dictionary.
4.  **Client**: Redraws the screen based on that dictionary.

## 1. The Server Code (Logic)

You have to run this in a separate `server.py` file, but here is the logic:

```python
players = {} # Mapping ID -> (x, y)

def threaded_client(conn, player_id):
    # 1. Send Initial Pos
    conn.send(pickle.dumps(players[player_id])) 
    
    while True:
        try:
            # 2. Receive Data
            data = pickle.loads(conn.recv(2048))
            players[player_id] = data # Update Server Memory

            # 3. Send Back EVERYONE'S Data
            conn.send(pickle.dumps(players))
        except:
            break
```
