## 1. Setting up a Socket Server

We use Python's built-in `socket` library. No install needed.

```python
import socket

# 1. Define the Machine
SERVER_IP = "localhost" # Or "127.0.0.1". This means "My Computer".
PORT = 5555             # The "Door Number" to knock on.

# 2. Create the Socket
# AF_INET = IPv4 (The Internet Address)
# SOCK_STREAM = TCP (Reliable, ordered data)
server = socket.socket(socket.AF_INET, socket.SOCK_STREAM)

# 3. Bind and Listen
try:
    server.bind((SERVER_IP, PORT))
    server.listen() # Open the ears!
    print(f"[SERVER] Waiting for connections on {SERVER_IP}:{PORT}...")
except socket.error as e:
    print(str(e))
```

## 2. Accepting Connections

The code above just "starts" the server. Now we need a loop to accept players when they connect.

```python
def start_server():
    while True:
        # The code halts here until someone connects!
        client, address = server.accept()
        print(f"[NEW CONNECTION] {address} connected.")
        
        # Send them a welcome message
        client.send(str.encode("Connected to Server!"))

# UNCOMMENT TO RUN (But warning: It hangs forever until broken!)
# start_server()
```
