## 1. Local Area Network (LAN) üè†

If your friend is in the **Same House** (connected to the same Wi-Fi), it's easy.

1.  **Find your Local IP Address**:
    *   Open Terminal (Cmd).
    *   Type `ipconfig` (Windows) or `ifconfig` (Mac/Linux).
    *   Look for `IPv4 Address`. It usually looks like `192.168.1.5`.

2.  **Update Server Code**:
    *   Change `SERVER_IP = "localhost"` to `SERVER_IP = ""` (This means "Listen to everyone").

3.  **Update Client Code**:
    *   Your friend must set `SERVER_IP = "192.168.1.5"` (Your specific IP).

Now you can play together in the same room!

## 2. The Internet (WAN) üåè

If your friend is in **Another House**, it's handled differently.

The Internet is blocked by your Router's Firewall. To let a friend in, you need **Port Forwarding**.

### Steps to Play Online:
1.  **Log into your Router** (Usually `192.168.0.1` in a browser).
2.  **Find "Port Forwarding"** settings.
3.  **Create a Rule**:
    *   Port: `5555` (The port we used in Python).
    *   IP: `192.168.1.x` (Your Computer's Local IP).
    *   Protocol: TCP.
4.  **Find your Public IP**:
    *   Go to Google and type: "What is my IP".
    *   Example: `45.12.89.201`.

### Connecting:
*   **You (Host)**: Run Server on `0.0.0.0` (Listen All).
*   **Friend (Client)**: Connects to `45.12.89.201` (Your Public IP).

Because Port Forwarding opens a literal hole in your security, **be careful who you share your IP with!**
