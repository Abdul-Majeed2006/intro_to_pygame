Sending text strings ("Hello") is boring.
We want to send **Game Data**:
```python
player_data = {
    "x": 100,

    "color": (255, 0, 0)
}
```

Sockets only understand **bytes**. So we need to "Pickle" our object (turn it into bytes) and "Unpickle" it on the other side.

## 1. Pickling Data

Python has a built-in library called `pickle`.

````python
import pickle

data = {"x": 50, "y": 50, "name": "Player 1"}

# TURN INTO BYTES
data_bytes = pickle.dumps(data)
print(f"Bytes: {data_bytes}")

# TURN BACK INTO OBJECT
restored_data = pickle.loads(data_bytes)
print(f"Restored: {restored_data}")
````

## 2. The Network Class

In a real game, you don't want to type `socket.send(pickle.dumps(data))` every time.
Let's make a Class that handles it all.

````python
import socket

class Network:
    def __init__(self):
        self.client = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
        self.server = "localhost"
        self.port = 5555
        self.addr = (self.server, self.port)
        
        self.p = self.connect()

    def connect(self):
        try:
            self.client.connect(self.addr)
            # Upon connection, maybe we receive our Starting Position?
            return pickle.loads(self.client.recv(2048))
        except:
            pass

    def send(self, data):
        try:
            self.client.send(pickle.dumps(data))
            return pickle.loads(self.client.recv(2048))
        except socket.error as e:
            print(e)
            return None
````
