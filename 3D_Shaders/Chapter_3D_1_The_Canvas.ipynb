## 1. The OpenGL Context

When we start Pygame, we need to tell the OS: "Give me a window, but let the GPU control it."

```python
import pygame
import moderngl
import sys

# 1. Initialize Pygame
pygame.init()

# 2. Tell Pygame to use OpenGL
# OPENGL: Use the GPU
# DOUBLEBUF: Use two screens (one for drawing, one for showing) to prevent flickering
pygame.display.set_mode((800, 600), pygame.OPENGL | pygame.DOUBLEBUF)

# 3. Create the Context (The handshake with the GPU)
ctx = moderngl.create_context()

# 4. Define the Viewport (Where are we drawing?)
ctx.viewport = (0, 0, 800, 600)
```

## 2. The Loop (Clear & Swap)

Instead of `screen.fill()`, we use `ctx.clear()`.
Instead of `pygame.display.flip()`, we use `pygame.display.flip()` (This part is the same!).

```python
running = True
while running:
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False
            
    # 1. Clear the Screen (Red, Green, Blue)
    # Typically we calculate 0.0 to 1.0 (float) instead of 0-255
    ctx.clear(0.2, 0.3, 0.3)
    
    # 2. Draw 3D Stuff?? (Next Chapter!)
    
    # 3. Swap Buffers
    pygame.display.flip()
    pygame.time.wait(16)

pygame.quit()
```
