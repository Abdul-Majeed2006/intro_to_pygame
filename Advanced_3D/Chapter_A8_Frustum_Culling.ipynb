Rendering 1 million triangles: fast.
Rendering 1 million triangles **you can't even see**: waste of GPU.

**Frustum culling** = Don't render what's outside the camera.
**Oct trees** = Organize space for fast queries.

This is how open-world games render huge maps at 60 FPS.
