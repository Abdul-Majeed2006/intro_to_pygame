Shadows make 3D scenes feel grounded in reality.

Without shadows, objects appear to float. With shadows, they feel **solid** and **present**.

This chapter teaches **Shadow Mapping**, the industry-standard technique for real-time shadows.

## 1. How Shadow Mapping Works

The algorithm has 2 passes:

### Pass 1: Render from Light's POV
- Render the scene from the light's position
- Store only **depth** (distance to light) in a texture
- This creates a "shadow map"

### Pass 2: Render from Camera's POV
- For each pixel, check: "Am I closer to the light than the shadow map says?"
- If YES ‚Üí lit
- If NO ‚Üí in shadow

## 2. Creating a Framebuffer (Render to Texture)

We need to render to a texture instead of the screen.

```python
# Create depth texture for shadow map
shadow_size = (2048, 2048)  # Higher = better quality
shadow_map = ctx.depth_texture(shadow_size)
shadow_map.filter = (moderngl.LINEAR, moderngl.LINEAR)

# Create framebuffer
shadow_fbo = ctx.framebuffer(depth_attachment=shadow_map)
```
