A character 1 meter away: render 10,000 triangles.
The same character 100 meters away: render 100 triangles.

**LOD** = Use simpler models for distant objects.

This is essential for open-world games with long draw distances.

## 1. Distance-Based LOD Switching

```python
class LODModel:
    def __init__(self):
        self.lod0 = load_model('character_high.obj')  # 10,000 tris
        self.lod1 = load_model('character_medium.obj')  # 2,000 tris
        self.lod2 = load_model('character_low.obj')  # 500 tris
        
    def get_lod_for_distance(self, distance):
        if distance < 10:
            return self.lod0
        elif distance < 50:
            return self.lod1
        else:
            return self.lod2
```
