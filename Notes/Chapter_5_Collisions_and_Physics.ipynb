## 1. Rectangle Collisions

The simplest collision: "Is this box touching that box?"

```python
if player.rect.colliderect(enemy.rect):
    print("OUCH! Hit by enemy!")
```

This works for single objects. But what if we have 50 bullets and 20 enemies?

## 2. Group Collisions (The Nuclear Option)

Pygame has a magic function: `pygame.sprite.groupcollide()`.

It checks two lists of sprites against each other and returns a dictionary of who hit who. 

```python
# Check if any BULLET hit any ENEMY
# True, True means "Kill the bullet AND Kill the enemy"
hits = pygame.sprite.groupcollide(bullets_group, enemies_group, True, True)

for bullet, enemies_hit in hits.items():
    score += 10
    print("Got one!")
```

## 3. Trigonometry (Homing Missiles) üß†

How does a missile know how to fly towards a moving target?
It calculates the **Angle** between them.

We need `math.atan2(dy, dx)`.

````python
import math

def get_target_angle(start_x, start_y, target_x, target_y):
    dx = target_x - start_x
    dy = target_y - start_y
    
    # Calculate angle in radians
    radians = math.atan2(dy, dx)
    return radians

def move_towards_target(x, y, angle, speed):
    # Convert angle back to X/Y velocity
    vx = math.cos(angle) * speed
    vy = math.sin(angle) * speed
    return x + vx, y + vy
````
