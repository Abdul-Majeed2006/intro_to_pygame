## 1. The Player Class with an ID

We need to tell the player: "Are you Player 1 or Player 2?"

```python
import pygame

class LocalPlayer(pygame.sprite.Sprite):
    def __init__(self, player_id, x, y, color):
        super().__init__()
        self.image = pygame.Surface((40, 40))
        self.image.fill(color)
        self.rect = self.image.get_rect(topleft=(x, y))
        self.id = player_id
        self.speed = 5
        
    def update(self):
        keys = pygame.key.get_pressed()
        
        # --- PLAYER 1 CONTROLS (WASD) ---
        if self.id == 1:
            if keys[pygame.K_w]: self.rect.y -= self.speed
            if keys[pygame.K_s]: self.rect.y += self.speed
            if keys[pygame.K_a]: self.rect.x -= self.speed
            if keys[pygame.K_d]: self.rect.x += self.speed
            
        # --- PLAYER 2 CONTROLS (ARROWS) ---
        elif self.id == 2:
            if keys[pygame.K_UP]: self.rect.y -= self.speed
            if keys[pygame.K_DOWN]: self.rect.y += self.speed
            if keys[pygame.K_LEFT]: self.rect.x -= self.speed
            if keys[pygame.K_RIGHT]: self.rect.x += self.speed
```

## 2. Adding Gamepads (The Real Co-Op Experience) üïπÔ∏è

If you plug in two USB controllers, Pygame sees them as `Joystick(0)` and `Joystick(1)`.

```python
    def update_with_joystick(self):
        # Get the specific controller for this player
        # Note: You need to initialize joysticks first in Main!
        if self.id <= pygame.joystick.get_count():
            # Player 1 gets Joystick 0, Player 2 gets Joystick 1
            pad = pygame.joystick.Joystick(self.id - 1)
            pad.init()
            
            # Get Axis (0=Left/Right, 1=Up/Down)
            dx = pad.get_axis(0)
            dy = pad.get_axis(1)
            
            # Deadzone (ignore tiny drifts)
            if abs(dx) > 0.1: self.rect.x += dx * self.speed
            if abs(dy) > 0.1: self.rect.y += dy * self.speed
```
