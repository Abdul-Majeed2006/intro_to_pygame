## 1. What is a Vector? üìê

In professional game dev, we don't handle `x` and `y` separately. We use a **Vector** which holds both.

pyGame has a built-in tool for this: `pygame.math.Vector2`.

```python
pos = pygame.math.Vector2(100, 200)
vel = pygame.math.Vector2(5, 0)

pos += vel # Automagically adds x to x and y to y!
```

## 2. The Physics Player (Professional Edition)

Let's build a clean Player class using Vectors.

````python
import pygame
vec = pygame.math.Vector2 # Alias for less typing

class Player(pygame.sprite.Sprite):
    def __init__(self):
        super().__init__()
        self.image = pygame.Surface((30, 50))
        self.image.fill((255, 255, 0))
        self.rect = self.image.get_rect()
        
        # --- POSITION & VELOCITY VECTORS ---
        # We use floating point numbers for smooth physics
        self.pos = vec(400, 300)
        self.vel = vec(0, 0)
        self.acc = vec(0, 0) # Acceleration
        
        # Config
        self.GRAVITY = 0.8
        self.JUMP_POWER = -15

    def jump(self):
        # Check if on ground before jumping
        self.rect.y += 1
        hits = pygame.sprite.spritecollide(self, platforms, False)
        self.rect.y -= 1
        if hits:
            self.vel.y = self.JUMP_POWER
            
    def update(self):
        # 1. Apply Gravity (Add to Acceleration Y)
        self.acc = vec(0, self.GRAVITY)
        
        # 2. Input (Add to Acceleration X)
        keys = pygame.key.get_pressed()
        if keys[pygame.K_LEFT]:
            self.acc.x = -0.5
        if keys[pygame.K_RIGHT]:
            self.acc.x = 0.5
            
        # 3. Apply Friction (Stopping when no key is pressed)
        self.acc.x += self.vel.x * -0.12
        
        # 4. Standard Physics Equation
        self.vel += self.acc
        self.pos += self.vel + 0.5 * self.acc
        
        # 5. Connect Physics back to Screen Position
        self.rect.midbottom = self.pos
        
        # 6. Safety: Don't fall forever
        if self.pos.y > 500: 
            self.pos.y = 500
            self.vel.y = 0
````

## 3. Floor Collision (The Hard Part)

When using Vectors, collision needs to snap (`position`) back to the top of the block.

````python
    def check_platform_collisions(self, platforms):
        if self.vel.y > 0: # Only check if falling down
            hits = pygame.sprite.spritecollide(self, platforms, False)
            if hits:
                # Snap to top of the first platform we hit
                self.pos.y = hits[0].rect.top
                self.vel.y = 0
                self.rect.midbottom = self.pos
````
