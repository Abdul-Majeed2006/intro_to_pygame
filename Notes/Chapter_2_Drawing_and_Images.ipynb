## 1. Loading an Image

Let's load the sand tile that makes up the floor of our desert.

We use `pygame.image.load(path)`. This loads the image into memory, but it doesn't draw it yet!

````python
pygame.init()
WIDTH, HEIGHT = 800, 600
screen = pygame.display.set_mode((WIDTH, HEIGHT))

# Load the image
sand_image_path = os.path.join(assets_path, "bg", "tile_0055.png")
sand_tile = pygame.image.load(sand_image_path)

print(f"Loaded image size: {sand_tile.get_width()} x {sand_tile.get_height()}")
````

## 2. To "Blit" or Not to "Blit"

"Blit" stands for **BLock Image Transfer**. It basically means: "Copy the pixels from this image and paste them onto the screen at these coordinates."

The coordinates `(0, 0)` are always at the **TOP LEFT** of the screen.

````python
running = True
while running:
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False
    
    # Draw the sand tile at (0, 0)
    screen.blit(sand_tile, (0, 0))
    
    # Draw another one at (100, 100)
    screen.blit(sand_tile, (100, 100))
    
    pygame.display.flip()

pygame.quit()
````

## 3. The Power of Rects üî≤

Every image has a `rect` (rectangle). This allows us to position images easily (e.g., `center`, `topright`, `midbottom`).

Instead of guessing (X, Y) numbers, we can say: "Put the player in the center of the screen."

````python
player_path = os.path.join(assets_path, "players", "tile_0006.png")
player_image = pygame.image.load(player_path)

# Get the rectangle of the image
player_rect = player_image.get_rect()

# Move the rectangle to the center of the screen
player_rect.center = (WIDTH // 2, HEIGHT // 2)

print(f"Player will be drawn at: {player_rect.x}, {player_rect.y}")
````
